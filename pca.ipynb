```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
wine = pd.read_csv('wine (1).csv')
```

```python
wine.head()
```

```python
from sklearn.preprocessing import scale
```

```python
scaled_df = scale(wine)
scaled_df
```

```python
#build model
```

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=13)
```

```python
pca_values = model.fit_transform(scaled_df)
```

```python
pca_values
```

```python
var=model.explained_variance_ratio_
var
```

```python
np.cumsum(var)
```

```python
final_model = PCA(n_components=3)
```

```python
pca_values = final_model.fit_transform(scaled_df)
```

```python
pca_values
```

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
clusters3=KMeans(3,random_state=30).fit(scaled_df)
clusters3
```

```python
clusters3.labels_
```
